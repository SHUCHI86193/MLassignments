```python
import numpy as np
import pandas as pd
import re
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import accuracy_score 
from sklearn.model_selection import train_test_split
```

```python
# Question 1
df_train = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv")
df_test = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv")
```

```python
data = pd.concat([df_train, df_test])
```

```python
data.head(5)
```

```python
# Question 2
data['Survived'] =  data['Survived'].dropna()
data['Survived'].value_counts()
```

```python
# Question 3
data['Has_Cabin'] = ~data.Cabin.isnull()
```

```python
# Question 5
data['Age'] = data.Age.fillna(data.Age.median())
data['Fare'] = data.Fare.fillna(data.Fare.median())
data['Embarked'] = data['Embarked'].fillna('S')
data.head(5)
```

```python
# Question 6
data['Categorical_age'] = pd.qcut(data.Age, q=2, labels=False )
data['Categorical_fare']= pd.qcut(data.Fare, q=2, labels=False)
data.head()
```

```python
# Question 7
data.drop(['Cabin', 'Name', 'PassengerId', 'Ticket', 'Age', 'Fare', 'SibSp', 'Parch'], axis=1, inplace=True)
```

```python
# Question 8
df = pd.get_dummies(data, drop_first=True)
```

```python
# Question 9
X = df.drop(['Survived'], axis = 1).values
y = df['Survived'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
# Question 10
m = linear_model.LogisticRegression()
m.fit(X_train, y_train)
predict = m.predict(X_test)
```

```python
# Question 11
acc_s = m.score(X_test, y_test)
print("Accuracy Score:{}".format(acc_s))
```
